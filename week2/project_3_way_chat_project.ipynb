```python
### Needs Improvement: This code is a great start for creating a multi-agent conversation using different language models. However, there are a few areas that could be improved for better functionality and readability:
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv()

# --------------------
# API KEYS
# --------------------

openai_api_key = os.getenv("OPENAI_API_KEY")
anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")

# --------------------
# CLIENTS
# --------------------

openai_client = OpenAI(api_key=openai_api_key)

anthropic_client = OpenAI(
    api_key=anthropic_api_key,
    base_url="https://api.anthropic.com/v1/"
)

ollama_client = OpenAI(
    api_key="ollama",
    base_url="http://localhost:11434/v1"
)

# --------------------
# AGENTS
# --------------------

agents = {
    "GPT": {
        "client": openai_client,
        "model": "gpt-4.1-mini",
        "system": (
            "You are argumentative and sarcastic. "
            "Keep replies short."
        )
    },

    "Claude": {
        "client": anthropic_client,
        "model": "claude-haiku-4-5",
        "system": (
            "You are polite and agreeable. "
            "Keep replies short."
        )
    },

    "Ollama": {
        "client": ollama_client,
        "model": "llama3.2",
        "system": (
            "You are philosophical and thoughtful. "
            "Keep replies short."
        )
    }
}

# --------------------
# SHARED CONVERSATION
# --------------------

conversation = [
    {
        "speaker": "User",
        "message": "Hi everyone!"
    }
]

# --------------------
# AGENT CALL
# --------------------

def call_agent(agent_name):

    agent = agents[agent_name]

    messages = [
        {
            "role": "system",
            "content": agent["system"]
        }
    ]

    # Convert shared conversation
    for entry in conversation:

        role = "user"

        if entry["speaker"] == agent_name:
            role = "assistant"

        messages.append({
            "role": role,
            "content": f'{entry["speaker"]}: {entry["message"]}'
        })

    response = agent["client"].chat.completions.create(
        model=agent["model"],
        messages=messages
    )

    return response.choices[0].message.content

# --------------------
# LOOP
# --------------------

turn_order = ["GPT", "Claude", "Ollama"]

for i in range(5):

    for agent_name in turn_order:

        reply = call_agent(agent_name)

        conversation.append({
            "speaker": agent_name,
            "message": reply
        })

        display(Markdown(
            f"### {agent_name}\n{reply}"
        ))
```
