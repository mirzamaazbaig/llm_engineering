````python
import sqlite3
````

````python
DB = "prices.db"

with sqlite3.connect(DB) as conn:
    cursor = conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS prices (city TEXT PRIMARY KEY, price REAL)')
    conn.commit()
````

````python
def set_ticket_price(city, price):
    with sqlite3.connect(DB) as conn:
        cursor = conn.cursor()
        cursor.execute('INSERT INTO prices (city, price) VALUES (?, ?) ON CONFLICT(city) DO UPDATE SET price = ?', (city.lower(), price, price))
        conn.commit()
````

````python
ticket_prices = {"london":799, "paris": 899, "tokyo": 1420, "sydney": 2999}
for city, price in ticket_prices.items():
    set_ticket_price(city, price)
````

````python
def get_ticket_price(city):
    print(f"DATABASE TOOL CALLED: Getting price for {city}", flush=True)
    with sqlite3.connect(DB) as conn:
        cursor = conn.cursor()
        cursor.execute('SELECT price FROM prices WHERE city = ?', (city.lower(),))
        result = cursor.fetchone()
        return f"Ticket price to {city} is ${result[0]}" if result else "No price data available for this city"
````

````python
get_ticket_price("Tokyo")
````

# 🗂️ SQLite CLI — How to View `prices.db` from Terminal

## 📍 Step 1: Open terminal in the correct folder
Make sure you are inside the directory where `prices.db` is located.

⚠️ Important:
If you are in the wrong folder, SQLite will **create a new empty database** instead of opening the existing one.

```bash
cd /path/to/folder/containing/prices.db

Write the following in the terminal **from the correct folder where `prices.db` is located**.

⚠️ If you don’t do this, the next step may create a **new empty database** instead of opening the existing one in your current directory.

```bash
sqlite3 prices.db
```

Then write:

```sql
.databases
```

If you see something like this:

```text
main: /path/to/prices.db
```

then you are working on the correct database.

Now you can run the following commands to view the data:

```sql
.tables
.headers on
.mode column
SELECT * FROM prices;
```
